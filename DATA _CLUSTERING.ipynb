# Knee Rule or Elbow method - to decide number of clusters

```python
from sklearn.cluster import KMeans

losses = []
for i in range(1,11):
    model1 = KMeans(n_clusters=i)
    model1.fit(df1)
    losses.append(model1.inertia_)
```

```python
print(losses)
```

```python
plt.plot(range(1,11) , losses)
plt.show()
```
